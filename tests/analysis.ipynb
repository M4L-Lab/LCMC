```python
from ase.io import read
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_energy_from_trajectory(trajectory):
    energies=[]
    for atoms in trajectory:
        energies.append(atoms.get_potential_energy())

    return np.array(energies)
```
